```python
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
%matplotlib notebook
```

```python
interval=100
circle_points=0
square_points=0
all_points=[]

for i in range(interval**2):
    rand_x = random.uniform(-1,1)
    rand_y = random.uniform(-1,1)
    origin_dist=rand_x**2+rand_y**2
    if origin_dist<=1:
        circle_points+=1
    square_points+=1
    all_points.append([rand_x, rand_y])
    pi=4*circle_points/square_points
    
all_points = np.array(all_points)
colors = np.array(["lightsteelblue" if i[0]**2 + i[1]**2 <= 1 else "lightslategrey" for i in all_points])

x = np.linspace(-1,1,100)
fig, ax = plt.subplots()
ax.plot(x, np.sqrt(1 - x**2), 'cadetblue')
ax.plot(x, -np.sqrt(1 - x**2), 'cadetblue')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
graph = ax.scatter([],[], marker='o', s=80)

def animate(i):
    graph.set_offsets(all_points[:i,:])
    graph.set_facecolor(colors[:i])
    return graph
    
animation = FuncAnimation(fig, func=animate, frames = range(np.shape(all_points)[0]), interval=5, repeat = False)
```
